```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Caracterización.xlsx')
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Genero', y='Edad', hue='Rol', data=df)  
plt.title('Distribución de la Edad por Género y Rol')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Genero', y='Años_experiencia', hue='Rol', data=df)
plt.title('Distribución de la Experiencia por Género y Rol')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours_Worked', y='Puntaje_productividad', hue='Rol', style='Genero', data=df)
plt.title('Productividad vs. Horas Trabajadas')
plt.xlabel('Horas Trabajadas')
plt.ylabel('Puntaje de Productividad')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Training_Hours', y='Puntaje_productividad', hue='Rol', style='Genero', data=df)
plt.title('Impacto de las Horas de Capacitación en la Productividad')
plt.xlabel('Horas de Capacitación')
plt.ylabel('Puntaje de Productividad')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Departamento', y='Puntaje_productividad', data=df, ci=None)
plt.title('Comparación de la Productividad por Departamento')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Puntaje_productividad'], bins=10, kde=True)
plt.title('Distribución del Puntaje de Productividad')
plt.xlabel('Puntaje de Productividad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Hours_Worked', y='Training_Hours', hue='Rol', style='Genero', data=df)
plt.title('Horas Trabajadas vs. Horas de Entrenamiento')
plt.xlabel('Horas Trabajadas')
plt.ylabel('Horas de Entrenamiento')
plt.show()
```
